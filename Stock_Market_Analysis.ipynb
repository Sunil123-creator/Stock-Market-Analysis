```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
#For rest of the ticker
appl = data["Adj Close"]["AAPL"]
amd = data["Adj Close"]["AMD"]
amzn = data["Adj Close"]["AMZN"]
jpm = data["Adj Close"]["JPM"]
meta = data["Adj Close"]["META"]
nvda = data["Adj Close"]["NVDA"]
spy = data["Adj Close"]["SPY"]
```

```python
adjt_close = data["Adj Close"]
anl_price = adjt_close.resample("Y").last()
anl_return = anl_price.pct_change().dropna()
anl_cumulative = (1 + anl_return).cumprod()
anl_cumulative
```

```python
import matplotlib.pyplot as plt
anl_cumulative.plot(figsize=(12,7))
plt.title("Annual Cumulative Return (Growth of $1)")
plt.ylabel("Portfolio Value ($)", fontsize=12)
plt.tight_layout()
plt.show()
```

### Insights:

In seven years, the best performing stock is NVDA ($1 to $56.99), followed by TSLA ($1 to $21.88) and AMD ($1 to $11.65). NVDA has rapid growth after 2023 which indicates strong AI-driven acceleration. TSLA had huge jump in 2019 to 2020 (10x). In 2022, all the stocks dropped sharply, especially TSLA. SPY and JPM are stable, slower growers whereas NVDA and TSLA have high growth and high volatility.

```python
#Lets calculate how each stock changes it's price
daily_return = adjt_close.pct_change().dropna()
daily_return.head(5)
```

```python
daily_volatility = daily_return.std()
daily_volatility
```

```python
#for annual voatility
import numpy as np
annual_volatility = daily_volatility * np.sqrt(252)
sorted_vol = annual_volatility.sort_values(ascending=False)
sorted_vol
```

```python
plt.figure(figsize=(12,7))
plt.bar(sorted_vol.index, sorted_vol.values, color="steelblue")
plt.title("Annual Volatility Comparison")
plt.ylabel("Annual Volatility")
plt.xlabel("Stock")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Insights:
The most volatile stock is TSLA (0.64) and the least volatile is SPY (0.19). AMD and NVDA are also among the high volatile stocks. Diversified market stock like JPM and SPY have lower volatility and greater stability.

```python
#return and risk
annual_return = daily_return.mean() * 252
```

```python
risk_return = pd.DataFrame({"Annual Return": annual_return, "Annual Volatility": annual_volatility})
risk_return
```

```python
plt.figure(figsize=(12,7))
plt.scatter( risk_return["Annual Volatility"], risk_return["Annual Return"], s=120 )

for ticker in risk_return.index:
    plt.text( risk_return.loc[ticker,"Annual Volatility"], risk_return.loc[ticker,"Annual Return"], ticker )

plt.title("Risk vs Return (Annual)")
plt.xlabel("Annual Volatility (Risk)")
plt.ylabel("Annual Return")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

### Insights:

Although technology stocks such as Tesla, Nvidia, and AMD have high volatility, however they also delievr higher returns. Whereas diversified stocks like JPM have low volatility and lower returns.

```python
#calculating sharpe ratio with risk free rate 2.5% (assumption)
risk_free_rate = 0.025
```

```python
sharpe_ratio = (annual_return - risk_free_rate) / (annual_volatility)
sort_sharpe = sharpe_ratio.sort_values(ascending=False)
sort_sharpe
```

```python
plt.figure(figsize=(12,7))
plt.bar(sort_sharpe.index, sort_sharpe.values, color="#4C72B0")
plt.title("Sharpe Ratio Comparison")
plt.ylabel("Sharpe Ratio")
plt.xlabel("Stock")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Insights:

NVIDIA has the highest Sharpe ratio and META has the lowest. NVIDIA has the best risk adjusted performace unlike AMD and TSLA which has the high return but also has the higher risk. Diversified investments such as SPY offer more stability but lower risk-adjusted performace compared with some individual growth stocks.


```python
summary = pd.DataFrame({"Annual Return": annual_return, "Annual Volatility": annual_volatility, "Sharpe Ratio": sharpe_ratio})
stock_summary = summary.sort_values("Sharpe Ratio", ascending=False)
stock_summary
```
